---
* This program is used for extracting and mapping ensemble mean data at each municipal from the subdirectory:
data\data\Future\EnsembleMean\Municipalities

* The original souce of the data is:
http://lamps.math.yorku.ca/OntarioClimate/index_app_data.htm#/StationsEnsembleMeanData

* The data are annual values at each weather station, for 45 variables from 1981-2099

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# user can change the variable name (select one from the varNames)
varNames=['Tm','Tx','Tn','Pr']
varName='Tx'
# csvfile name is:
rcp26csvfile = '../data/data/Future/EnsembleMean/Municipalities/' + varName + '_150stations_RCP26_EnsembleMean_Ann_head.csv'
rcp45csvfile = '../data/data/Future/EnsembleMean/Municipalities/' + varName + '_150stations_RCP45_EnsembleMean_Ann_head.csv'
rcp60csvfile = '../data/data/Future/EnsembleMean/Municipalities/' + varName + '_150stations_RCP60_EnsembleMean_Ann_head.csv'
rcp85csvfile = '../data/data/Future/EnsembleMean/Municipalities/' + varName + '_150stations_RCP85_EnsembleMean_Ann_head.csv'

data26 = pd.read_csv(rcp26csvfile)
data45 = pd.read_csv(rcp45csvfile)
data60 = pd.read_csv(rcp60csvfile)
data85 = pd.read_csv(rcp85csvfile)

# print(data26.head())
# cityNames=data0['RegName']
# print(cityNames)
cityName ='AMHERSTBURG'

data26 = pd.read_csv(rcp26csvfile)
rcp26=np.array(data26[data26['RegName']==cityName])[0,2:]
print(np.shape(rcp26))
rcp45=np.array(data45[data45['RegName']==cityName])[0,2:]
rcp60=np.array(data60[data60['RegName']==cityName])[0,2:]
rcp85=np.array(data85[data85['RegName']==cityName])[0,2:]
year=np.arange(1981,2100)
```

```python
figure=plt.figure(figsize=(18,10))
l1=plt.plot(year,rcp26,color='blue')
l2=plt.plot(year,rcp45,color='c')
l3=plt.plot(year,rcp60,color='orange')
l4=plt.plot(year,rcp85,color='red')
plt.legend(('RCP2.6','RCP4.5','RCP6.0','RCP8.5'))
plt.xlabel('Year')
plt.ylabel(varName)
titletext=cityName 
plt.title(titletext)
```
