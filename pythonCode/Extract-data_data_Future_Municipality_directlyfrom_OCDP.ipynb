```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
dir0 = 'http://lamps.math.yorku.ca/OntarioClimate/DS_CMIP5/LAMPS_YorkU/Stations/'
```

```python
city = 'AJAX'                               ## one of the 150 cities
varName = 'tasmax'                             ## tas, tasmax, tasmin or pr
institute = 'LAMPS'                        ## LAMPS or CORDEX
rcp = 'RCP26'                               ## RCP26, RCP45, RCP60 or RCP85
# be careful to revised the following code
if institute == 'CORDEX': 
    gcm = 'RCP26_0.44_RCA4_EC_EARTH_365X120'    ## check the following list, and revise according to your requirment
    url = dir0 + city + '/' + varName + '/' + rcp + '/' + varName + '_' + institute + '_' + gcm + '.csv'
print(url)
# we can also directly copy full url from OCDP website to url for example
 # url='http://lamps.math.yorku.ca/OntarioClimate/DS_CMIP5/LAMPS_YorkU/Stations/AJAX/tas/RCP85/tas_LAMPS_ACCESS1-0_RCP85.csv'
```

```python
rcp='RCP85'
if institute == 'LAMPS':
    gcm = 'ACCESS1-0_RCP85'
    url = dir0 + city + '/' + varName + '/' + rcp + '/' + varName + '_' + institute + '_' + gcm + '_365X120.csv'
print(url)
# we can also directly copy url from OCDP website to url for example
 # url='http://lamps.math.yorku.ca/OntarioClimate/DS_CMIP5/LAMPS_YorkU/Stations/AJAX/tas/RCP85/tas_LAMPS_ACCESS1-0_RCP85_365X120.csv'
```

```python
# url='AJAX/tas/RCP26/tas_CORDEX_RCP26_0.44_RCA4_EC_EARTH_365X120.csv'
data = pd.read_csv(url,sep=",", header=None) # use sep="," for coma separation. 
print(data.shape)
# 365 rows(day),120 column(years: 1981-2100)
print(data.head(1))
data.describe()
```

```python
fig2 = plt.figure(figsize=(18,18))
ax1=plt.subplot(5,1,1)
timeseriesData=np.array(data)
plt.plot(timeseriesData.T.flatten())
Nyear=120
Nday=365
# following we generate date range and calculate monthly mean, annuan mean
annualMean = np.mean(timeseriesData,axis=0)
Yrs = np.arange(Nyear)+1981
ax3=plt.subplot(5,1,3)
plt.plot(Yrs,annualMean)
start=datetime.datetime.strptime('01-01-1981','%d-%m-%Y')
end = datetime.datetime.strptime('01-01-2101','%d-%m-%Y')
dates = [start + datetime.timedelta(days=x) for x in range(0,(end-start).days)]
# delete February 29
dates1 = [x for x in dates if not ((x.month == 2) and (x.day == 29))]

df_data = pd.DataFrame(dates1,columns =['date'])
df_data['year'] = [x.year for x in dates1]
df_data['monthinyear'] = [x.month for x in dates1]
df_data['dayinmonth'] = [x.day for x in dates1]

df_data['weekinyear'] = [x.isocalendar()[1] for x in dates1]
df_data['dayinweek'] = [x.isocalendar()[2] for x in dates1]
df_data['values'] = timeseriesData.T.flatten()
# print(df_data.head(40))

# df_data.to_csv('replace with your path where you want to store the csv file/wantedcsvfilename.csv')

monthlyMean = df_data.groupby(['year','monthinyear'],as_index=False).agg({"values": "mean"})
# extract 30-yrs data 1981-2010
data30Yrs = df_data[df_data['year'] > 1980]

data30Yrs = data30Yrs[data30Yrs['year'] < 2011]                      
print(np.shape(data30Yrs))

# monthly30YrsMean = df_data.group
# print(monthlyMean.head(10)) 
ax2=plt.subplot(5,1,2)
plt.plot(monthlyMean.year+(monthlyMean.monthinyear-1)/12,monthlyMean['values']) 

ax1.title.set_text('Daily')
ax2.title.set_text('Monthly')
ax3.title.set_text('Annual')
```
